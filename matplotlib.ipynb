# matplotlib- Python Librery

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,45,6,67,8]
y=[13,15,17,9,10,12]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='solid')
```

```python
plt.plot(x,y,'g',linestyle='',marker='+',markersize='8')
```

```python
plt.plot(x,y,'g',alpha=0.5)

### alpha will set its tranparancy between 0 and 1.
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

# Bar Chart

```python
company=['GOOGLE','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python

import numpy as np
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.bar(ypos,revenue)

### or we can do directly plt.plt(company,revenue)

plt.ylabel('revenue(bln)')
plt.title("US Tech Stocks")

plt.bar(ypos,revenue,label='Revenue')
plt.bar(ypos,profit,label='profit')
plt.legend()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos-0.06,revenue,label='Revenue')
plt.bar(xpos+0.06,profit,label='profit')
plt.legend()

### to design side by side bar 
```

```python
### or we can do this in another way

plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')
plt.legend()

### to design side by side bar 
```

***Horizontal Bar Chart***

```python
plt.yticks(ypos,company)
plt.title("US Tech Stocks")
plt.barh(xpos-0.02,revenue,label='Revenue')
plt.barh(xpos+0.02,profit,label='profit')
plt.legend()
```

# Histogram

```python
blood_suger=[113,85,90,150,140,88,93,115,135,80,77,82,129]

plt.hist(blood_suger, bins=3, rwidth=0.95, color='red')

###In a histogram, bins are used to group data into intervals or "bins" and display the frequency or count of data points that fall into each bin. The x-axis of a histogram represents the data range, divided into these bins, and the y-axis represents the frequency or count of data points that belong to each bin.
### rwidth is a relative width of bins
```

```python
### to plot to histogram side by side

blood_suger=[113,85,90,150,140,88,93,115,135,80,77,82,129]
blood_suger_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist([blood_suger,blood_suger_women], bins=3, rwidth=0.95, color=['red','green'],orientation='horizontal',label=['men','women'])
plt.xlabel('Suger range')
plt.ylabel('Total no of patient')
plt.title('Blood suger analysis')
plt.legend()
### if we want to print horizontal histogram then...orientation='horizontal
```

# Pie Chart

```python
exp_vals=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill','car','Other Utilites']
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.1f%%',explode=[0,0.1,0,0,0])
plt.show()

### use autopct to show the values in chart.
###use explode th show that particular range is coming out of the rest of the chart.
### also we can use startangle=180 or 45 or 90 to set particular direction.

plt.savefig('piechart.png',bbox_inches='tight',pad_inches=2,transparent=True)

### by using these methods we can modify the layout of our file where we save this.
### using bbox_inches=tight it will contain whole chart tightly in saved file. we can give extra space around this chart in our file by giving the method like pad_inches and set the values for it
### using transparent property we can generate tranparent file of it.
### we can save the chart in pdf file instead of png file and we can give the location where we want to save it.
```
